FUNCIONES DENTRO DE FUNCIONES 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x):
  return x**3

def f(x):
  return np.sin(x)

x = np.linspace(-10,10, num=1000)

f_o_g = f(g(x))

plt.plot(x,f_o_g)
plt.grid()
```
