# MANIPULACION DE FUNCIONES 

```python
import matplotlib.pyplot as plt   # librería para graficar
import numpy as np                # librería para manejo de vectores y utilidades matemáticas
```

DESPLAZAMIENTOS 

```python
N = 100 #numero de puntos, primer dato a considerar 

def f(x):
  return x**2;
   
c = 10

x = np.linspace(-5, 5, num=N)

b = 5

# f(x) + c hacia arriba
# f(x) - c hacia abajo
# f(x-c) hacia la derecha
# f(x-c) hacia la izquierda

y = f(x +c ) 


fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

ALEJAMIENTO Y COMPRESIONES 

```python
def f(x):
  return np.sin(x);
   
c = 2

x = np.linspace(-15, 15, num=N)


# c * f(x)  alarga verticalmente
# 1/c * f(x) comprime verticalmente
# f(c*x)     comprime horizontalmente 
# f(1/c *x ) alarga horizontalmente 

y = f(x)


fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```

REFLEXIONES 

```python
def f(x):
  return x**3;
   
c = 2

x = np.linspace(-10, 10, num=N)


# - f(x)  refleja la grafica respecto al eje x
# f(-x)   refleja la grafica respecto al aje y 
 

y = f(-x)


fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```
