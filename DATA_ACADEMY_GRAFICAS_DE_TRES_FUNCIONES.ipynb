GRAFICO DE TRES DIMENSIONES 

```python
from matplotlib import cm #manejo de colores en grafica
import numpy as np
import matplotlib.pyplot as plt
```

```python

res = 100

Z = np.sin(X) + 2*np.sin(Y) 

X = np.linspace(-4, 4, res)
Y = np.linspace(-4, 4, res)

X, Y = np.meshgrid(X, Y)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

surf = ax.plot_surface(X, Y, Z, cmap=cm.cool,
                       linewidth=0, antialiased=False)

fig.colorbar(surf)
```

CURVAS DE NIVEL

```python
fig2, ax2 = plt.subplots()
level_map = np.linspace(np.min(Z),np.max(Z), num =res)

cp = ax2.contour(X,Y,Z,levels=level_map,cmap=cm.cool)
fig2.colorbar(cp)
```

GRAFICA  3D DE LA FUNCION DE COSTE 

```python
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})

def f(x,y):
  return x**2 + y**2

x = np.linspace(-4,4, res)
y = np.linspace(-4,4, res)

x,y = np.meshgrid(x,y)

z = f(x,y)

surf  = ax.plot_surface(x,y,z, cmap=cm.cool)

fig.colorbar(surf)
```

DESCENSO DE GRADIENTE

```python
level_map =np.linspace(np.min(z), np.max(z), res)
plt.contourf(x,y,z, levels=level_map, cmap =cm.cool)
plt.colorbar()
plt.title("descenso del gradiente")

p = np.random.rand(2) * 8 - 4 
plt.plot(p[0], p[1], 'o',c = 'k')

lr = 0.01 
h = 0.01

def derivate(cp, p):
  return (f(cp[0], cp[1]) - f(p[0], p[1])) / h

def gradient(p):
  grad = np.zeros(2)
  for idx, val in enumerate(p):
    cp = np.copy(p)
    cp[idx] = cp[idx] + h 

    dp = derivate (cp, p)
    grad[idx] = dp 

  return grad

for i in range(1000):
  p =  p  - lr*gradient(p)
  if(i % 10 == 0 ):
    plt.plot(p[0], p[1], 'o',c = 'r')
  
plt.plot(p[0], p[1], 'o',c = 'w')
print(p)
```
